# Input

```python
# Import csv data
import pandas as pd
from tqdm import tqdm
df = pd.read_json('contacts.json')
```

```python
# Check several data
pd.set_option('display.max_colwidth', None)
df.head()
```

# ???

```python
df['Partner'] = -1
```

```python
df = df.sort_values(by=['Email'], ascending=False).reset_index(drop=True)
for i in tqdm(range(len(df))):
    if i > 0:
        if df.loc[i, 'Email'] != '' and df.loc[i, 'Email'] == df.loc[i-1, 'Email']:
            if df.loc[i, 'Partner'] == -1:
                df.at[i, 'Partner'] = df.loc[i-1, 'Id']
            if df.loc[i-1, 'Partner'] == -1:
                df.at[i-1, 'Partner'] = df.loc[i, 'Id']
```

```python
df = df.sort_values(by=['Phone'], ascending=False).reset_index(drop=True)
for i in tqdm(range(len(df))):
    if i > 0:
        if df.loc[i, 'Phone'] != '' and df.loc[i, 'Phone'] == df.loc[i-1, 'Phone']:
            if df.loc[i, 'Partner'] == -1:
                df.at[i, 'Partner'] = df.loc[i-1, 'Id']
            if df.loc[i-1, 'Partner'] == -1:
                df.at[i-1, 'Partner'] = df.loc[i, 'Id']
```

```python
df = df.sort_values(by=['OrderId'], ascending=False).reset_index(drop=True)
for i in tqdm(range(len(df))):
    if i > 0:
        if df.loc[i, 'OrderId'] != '' and df.loc[i, 'OrderId'] == df.loc[i-1, 'OrderId']:
            if df.loc[i, 'Partner'] == -1:
                df.at[i, 'Partner'] = df.loc[i-1, 'Id']
            if df.loc[i-1, 'Partner'] == -1:
                df.at[i-1, 'Partner'] = df.loc[i, 'Id']
```
